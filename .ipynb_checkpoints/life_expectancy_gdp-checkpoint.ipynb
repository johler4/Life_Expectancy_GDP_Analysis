## Basic Analysis Questions
1) What is the overall trend in life expectency over the years?

3) Is there a correlation between life expectancy and gdp for the entire dataset?

```python
# 1) What is the overall trend in life expectency over the years?
sns.scatterplot(x='Year', y='Life_expectancy', hue='Country', data=data)
plt.title('Life Expectancy over Years')
plt.xlabel('Years')
plt.ylabel('Life Expectancy')
plt.show()
plt.clf()
```

```python
# 2) Are there any countries with consistently increasing life expectancy throughout the years?
sns.lineplot(x='Year', y='Life_expectancy', hue='Country', data=data)
plt.title('Life Expectancy over Years')
plt.xlabel('Years')
plt.ylabel('Life Expectancy')
plt.show()
plt.clf()
```

```python
# 3) Is there a correlation between life expectancy and GDP for the entire dataset?
sns.scatterplot(x='Life_expectancy', y='GDP', data=data)
plt.title('Life Expectancy Compared to GDP')
plt.xlabel('Life Expectancy')
plt.show()
plt.clf()
```

## Intermediate Analysis Questions
1) Can you identify any outliers in the gdp column using visualizations like box plots or scatter plots?
2) What is the average life expectancy for each country, and how does it vary over time?

```python
# 1) Can you identify any outliers in the GDP column using visualizations like box plots or scatter plots?
plt.figure(figsize=(4, 10))
sns.boxplot(y='GDP', data=data)
plt.title('Summary Statistics: Overall GDP\'s')
plt.show()
plt.clf()
```

> The boxplot above show many outliers that go beyond the "maximum" point for the overall statistics of the GDP's in our dataset.

```python
2) What is the average life expectancy for each country, and how does it vary over time?
for country in data.Country.unique():
    start_value = data.Life_expectancy[(data.Country == country) & (data.Year == 2000)].item()
    end_value = data.Life_expectancy[(data.Country == country) & (data.Year == 2015)].item()
    diff= end_value - start_value
    print(f'The average life expectancy for {country} between 2000-2015 is {data.Life_expectancy[data.Country == country].mean()}.')
    print(f'The average life expectancy changed by {round(diff, 2)} from the year 2000 to 2015 in {country}.')

sns.lineplot(x='Year', y='Life_expectancy', hue='Country', data = data)
plt.show()
plt.clf()

# sns.boxplot(x='Country', y='Life_expectancy', data=data)
# plt.show()
# plt.clf()
```

> In the 15 years worth of data that we observed, Germany had the highest average life expectancy and Zimbabwe had the lowest. That being said, Zimbabwe's life expectancy grew by the largest amount; a 14.7 year difference from its amount in 2000 compared to its amount in 2015. The other countries only show a change of life expectancy in the range of 1.9-4.4 years.

## Advanced Analysis Questions
1) Using a regression model, can you predict life expectency base on GDP for a specific country?
2) Are there any countries that experienced a sudden increase in GDP and how did it affect life expectancy?
3) Create a choropleth map to visualize the life expectancy values on a world map.

```python
# 1) Using a regression model, can you predict life expectancy based on GDP for a specific country?
# we have a dataset named 'data' with columns GDP and Life_expectancy. If we want to fit a model that can predict \
# life expectancy based on GDP, we can create the model (specify only one country for the model):
import statsmodels.api as sm
model = sm.OLS.from_formula('Life_expectancy ~ GDP', data=data[data.Country == 'Chile'])
results = model.fit()

# print(results.params)
# y = mx + b
# life_expectancy = results.params[1] * GDP + results.params[0]

random_gdp = 10.9854215e+10
test_life_exp = results.params[1] * random_gdp + results.params[0]
print(f'The life expectancy for Chile if its GDP were {"{:,}".format(random_gdp)} is expected to be {round(test_life_exp, 2)}.')
```

> When fitting the information about life expectancy with GDP for Chile in a regression model, we can predict that with an X GDP value, Chile's life expectancy will be Y. For a specific example, with a GDP of $10.9854215e+10, Chile would have a life expectancy of 78.16.

```python
# 2) Are there any countries that experienced a sudden increase in GDP and how did it affect life expectancy?
# calculate year to year change of GDP for each country and add column to original dataframe of results
data['GDP_change'] = data.groupby('Country')['GDP'].diff()
data['Life_expectancy_change'] = data.groupby('Country')['Life_expectancy'].diff()
# print(data.head(20))

# # testing formulas
# print(data.GDP_change[(data.Country == 'Chile') & (data.Year == 2001)].item() / data.GDP[(data.Country == 'Chile') & (data.Year == 2001)].item())

data['GDP_change'] = data.GDP_change / data.GDP
# print(data.head(10))

# define a threshold of significant change
threshold = 0.20

# filter countries with changes in gdp that are >= / < threshold
lesser_threshold_data = data[data.GDP_change < threshold]
greater_threshold_data = data[data.GDP_change >= threshold]

# create list of countries that have increases in GDP >= threshold
greater_threshold_countries = greater_threshold_data.Country.unique()
print(f'Countries that experienced sudden increases in GDP: {greater_threshold_countries}')

print(f'Average change in life expectancy per year for all data: {data.Life_expectancy_change.mean()}')
print(f'Average change in life expectancy for years that had a sudden increase in GDP: {round(greater_threshold_data.Life_expectancy_change.mean(), 2)}')
print(f'Average change in life expectancy for years that did not have a sudden increase in GDP: {round(lesser_threshold_data.Life_expectancy_change.mean(), 2)}')

sns.boxplot(x='Country', y='Life_expectancy_change', data=data)
plt.xticks(rotation=20)
plt.show()
plt.clf()
```

> The average change in life expectancy from year to year for all data is an increase of 0.33. Chile, China, and Zimbabwe all experienced years with a significant increase in GDP (20% or greater). Among the data of significant GDP increase, the average increase in life expectancy was 0.42 per year. The average increase of life expectancy among the data of insignificant GDP increase was 0.32; much closer to the overall average. This data shows us that significant increase in GDP may correlate to larger increases in life expectancy.

```python
# 3) Create a choropleth map to visualize the life expectancy values on a world map.
# import plotly and plotly.express
import plotly
import plotly.express as px
# create fig
fig = px.choropleth(data, locations='Country', locationmode='country names', color='Life_expectancy')
fig.show()
```

## Hypothesis Testing and Advanced Visualizations
1) Create a heat map to visualize the correlation matrix between life expectancy, GDP, and year.
2) Investigate whether there is a time lag between changes in GDP and changes in life expectancy for specific countries.

```python
# 1) Create a heat map to visualize the correlation matrix between life expectancy, GDP, and Year.
data2 = data
GDP_life_exp_year = data2.reset_index().pivot(index='Year', columns='Country', values='Life_expectancy')

# create heatmap
sns.heatmap(GDP_life_exp_year)
plt.title('Life Expectancy per Country based on Year')
plt.show()
plt.clf()
```

```python
# 2) Investigate whether there is a time lag between changes in GDP and changes in life expectancy for specific countries.
# import module
from pandas.plotting import lag_plot
lag_plot(data.Life_expectancy_change)
plt.title('Lag Plot of Life Expectancy Changes')
plt.show()
plt.clf()

lag_plot(data.GDP_change)
plt.title('Lag Plot of GDP Changes')
plt.show()
plt.clf()
```

> Both charts show points moving from bottom left to top right, indicating that there is a positive correlation between observations and their lag times for both GDP and life expectancy.
