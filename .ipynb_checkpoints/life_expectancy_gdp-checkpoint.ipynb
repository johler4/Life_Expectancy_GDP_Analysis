```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('all_data.csv')
print(data.head())
```

```python
# rename life expectancy column to make it easier to write code for later
data = data.rename(columns=({'Life expectancy at birth (years)':'Life_expectancy'}))
print(data.head())
```

```python
# 1) What are the dimensions of the dataset?
# 2) What are the data types of each columns in the dataset?
print(data.info())
```

```python
# 4) What is the time range of the data?
print(f'The years range from {min(data.Year)} to {max(data.Year)}.')
```

```python
# 5) Provide summary statistics for life expectancy and GDP.
print(data.Life_expectancy.describe())
print(data.GDP.describe())
```

## Basic Analysis Questions
1) What is the overall trend in life expectency over the years?
2) Are there any countries with consistently increasing life expectency throughout the years?
3) Is there a correlation between life expectancy and gdp for the entire dataset?

```python
# 1) What is the overall trend in life expectency over the years?
sns.scatterplot(x='Year', y='Life_expectancy', hue='Country', data=data)
plt.title('Life Expectancy over Years')
plt.xlabel('Years')
plt.ylabel('Life Expectancy')
plt.show()
plt.clf()
```

> The overall trend for life expectancy over the years is positive, with some countries having larger growth curves than others.

```python
# 2) Are there any countries with consistently increasing life expectancy throughout the years?
sns.lineplot(x='Year', y='Life_expectancy', hue='Country', data=data)
plt.title('Life Expectancy over Years')
plt.xlabel('Years')
plt.ylabel('Life Expectancy')
plt.show()
plt.clf()
```

> All countries except for Zimbabwe showed consistent growth from years 2000-2014.Zimbabwe actually had a decrease in life expectancy from 2000-2004 and then had rapid growth from 2005-2014.

```python
# 3) Is there a correlation between life expectancy and GDP for the entire dataset?
sns.scatterplot(x='Life_expectancy', y='GDP', data=data)
plt.title('Life Expectancy Compared to GDP')
plt.xlabel('Life Expectancy')
plt.show()
plt.clf()
```

```python
# separate scatterplot for zimbabwe's outlieing lower life expectancies
zimbabwe = data[data.Country == 'Zimbabwe']
sns.scatterplot(x='Life_expectancy', y='GDP', data=zimbabwe)
plt.title('Zimbabwe\'s Life Expectancy VS GDP')
plt.xlabel('Life Expectancy')
plt.ylabel('GDP')
plt.show()
plt.clf()
```

> The scatter plots of Life_expectancy against GDP show that there is a positive linear correlation between the two variables.

## Intermediate Analysis Questions
1) Can you identify any outliers in the gdp column using visualizations like box plots or scatter plots?
2) What is the average life expectancy for each country, and how does it vary over time?

```python
# 1) Can you identify any outliers in the GDP column using visualizations like box plots or scatter plots?
sns.boxplot(y='GDP', data=data)
plt.title('Summary Statistics: Overall GDP\'s')
plt.show()
plt.clf()
```

> The boxplot above show many outliers that go beyond the "maximum" point for the overall statistics of the GDP's in our dataset.

```python
# 2) What is the average life expectancy for each country, and how does it vary over time?
for country in data.Country.unique():
    print(f'The average life expectancy for {country} between 2000-2015 is {data.Life_expectancy[data.Country == country].mean()}.')

sns.lineplot(x='Year', y='Life_expectancy', hue='Country', data = data)
plt.show()
plt.clf()


# sns.boxplot(x='Country', y='Life_expectancy', data=data)
# plt.show()
# plt.clf()
```
